## Advance Operation with Business Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [resturant_id, 2021, 2022, 2023, 2024 ]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [5, 160000, 185000, 205000, 230000], #Chai Point
])
```

### === Zameto sales analysis ===

```python
#Shape of given array
print("\nShape of sales data array : ", sales_data.shape)

#View sample data
print("\nSample data for 1st three Resturant :", sales_data[:3])

#Total sales per year
print("\nTotal sales per year of resturant :", np.sum(sales_data[:,1:], axis = 0))

#Minimum sales per resturant
min_sales = np.min(sales_data[:, 1:] , axis = 1)
print("\nMinimum sales per resturant :", min_sales)

#Maximum sales per year
max_sales = np.max(sales_data[:, 1:] , axis = 0)
print("\nMaximum sales per year :", max_sales)

#Average sales per resturent
avg_sales = np.mean(sales_data[:, 1:] , axis = 1)
print("\nAverage sales per resturant :", avg_sales)

```

### Little help of Matplot 

```python
#Cumulative sales 
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sales:", cumsum)
plt.figure(figsize=(10, 6))

# Mean cumulative sales per restaurant
plt.plot(np.mean(cumsum, axis=1))

plt.title("Average cumulative sales per restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")

plt.grid(True)
plt.show()
```

## Vector Operations

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

#Vector Addition
print("\nVector Addition :", vector1 + vector2 )

#Vector Multiplication 
print("\nMultiplication of Vector:", vector1 * vector2 )

#Dot product
print("\nDot Product:", np.dot(vector1, vector2))

#Angle of Vector
print("\nAngle between vectors :", np.arccos(np.dot(vector1, vector2)/ (np.linalg.norm(vector1)* np.linalg.norm(vector2))))
# np.linalg.norm() computes the magnitude (length) of a vector
```

## Vectorized operation

```python
restaurant_types = np.array(["biryani", "chinese", "pizza", "burger", "cafe"])

#np.vectorize() applies a Python function to each element of an array
vectorized_upper = np.vectorize(str.upper)
print("Vectorize upper", vectorized_upper(restaurant_types))
```
